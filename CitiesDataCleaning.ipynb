```python
#IMPORTATION DES LIBRAIRIES
import os
import pandas as pd
```

```python
#CHARGEMENT DES DONNEES 
df_ville = pd.read_csv("Cities.csv", sep=",")

#SUPPRESSION DES COLONNES INUTILES
df_ville = df_ville.drop(columns=["insee_code", "city_code", "label", "department_name", "region_name"])
df_ville.head()
```

```python
df_caf = pd.read_csv("ndur_s_qf_400_dep.csv", sep=";")
df_caf.head()
```

```python
#SUPPRESSION DES VALEURS MANQUANTES
df_ville = df_ville.dropna()
```

```python
#Renommer les colonnes
df_caf.rename(columns={"Date référence": "date", 
                                "Numéro département": "num_dep", 
                                "Nom département": "nom_dep", 
                                "Numéro région": "num_region", 
                                "Nom région": "nom_region", 
                                "Quotient familial": "qt_familial", 
                                "Nombre foyers NDUR": "nb_foyers_presta_legale", 
                                "Nombre personnes NDUR": "nb_pers_presta_legale", 
                                "Montant total NDUR": "mt_total_presta_legale",
                                "Nombre foyers NDURPAJE": "nb_foyers_petite_enfance",
                                "Nombre personnes NDURPAJE": "nb_pers_petite_enfance",
                                "Montant total NDURPAJE": "mt_total_petite_enfance",
                                "Nombre foyers NDUREJ": "nb_foyers_enfance_jeunesse",
                                "Nombre personnes NDUREJ": "nb_pers_enfance_jeunesse",
                                "Montant total NDUREJ": "mt_total_enfance_jeunesse",
                                "Nombre foyers NDURAL": "nb_foyers_apl",
                                "Nombre personnes NDURAL": "nb_pers_apl",
                                "Montant total NDURAL": "mt_total_apl",
                                "Nombre foyers NDURINS": "nb_foyers_solidarite_insertion",
                                "Nombre personnes NDURINS": "nb_pers_solidarite_insertion",
                                "Montant total NDURINS": "mt_total_solidarite_insertion",
                                "Lieu résidence": "lieu_residence"}, inplace=True)
```
