```python
#IMPORTATION DES LIBRAIRIES
import os
import pandas as pd
```

```python
#CHARGEMENT DES DONNEES 
df_ville = pd.read_csv("Cities.csv", sep=",")

#SUPPRESSION DES COLONNES INUTILES
df_ville = df_ville.drop(columns=["insee_code", "city_code", "label", "region_name"])
dt_ville = pd.DataFrame(df_ville)

dt_ville.head()
```

```python
dt_ville['department_name'] = dt_ville['department_name'].str.upper()
dt_ville.head()
```

```python
#CHARGEMENT DES DONNEES DE LA BASE DONNEES DE LA CAF
df_caf = pd.read_csv("ndur_s_qf_400_dep.csv", sep=";")
df_caf.head()
```

```python
#CHECK FOR MISSING VALUES
print(dt_ville.isnull().sum())
```

```python
print(df_caf.isnull().sum())
```

**La base de données *VILLES(df_villes)* contient 211 données de Latitude et Longitude manquantes**  \
**La base de données *CAF(df_caf)* contient 1410 noms de départements manquants et 2186 nom de région et lieu de résidence manquants**

J'ai supprimé les données manquantes de la base VILLES mais j'ai gardé les données manquantes de la base de données CAF

```python
#SUPPRESSION DES VALEURS MANQUANTES
dt_ville = dt_ville.dropna()
```

```python
#Renommer les colonnes de la base de données de la CAF
df_caf.rename(columns={"Date référence": "date", 
                                "Numéro département": "num_dep", 
                                "Nom département": "nom_dep", 
                                "Numéro région": "num_region", 
                                "Nom région": "nom_region", 
                                "Quotient familial": "qt_familial", 
                                "Nombre foyers NDUR": "nb_foyers_presta_legale", 
                                "Nombre personnes NDUR": "nb_pers_presta_legale", 
                                "Montant total NDUR": "mt_total_presta_legale",
                                "Nombre foyers NDURPAJE": "nb_foyers_petite_enfance",
                                "Nombre personnes NDURPAJE": "nb_pers_petite_enfance",
                                "Montant total NDURPAJE": "mt_total_petite_enfance",
                                "Nombre foyers NDUREJ": "nb_foyers_enfance_jeunesse",
                                "Nombre personnes NDUREJ": "nb_pers_enfance_jeunesse",
                                "Montant total NDUREJ": "mt_total_enfance_jeunesse",
                                "Nombre foyers NDURAL": "nb_foyers_apl",
                                "Nombre personnes NDURAL": "nb_pers_apl",
                                "Montant total NDURAL": "mt_total_apl",
                                "Nombre foyers NDURINS": "nb_foyers_solidarite_insertion",
                                "Nombre personnes NDURINS": "nb_pers_solidarite_insertion",
                                "Montant total NDURINS": "mt_total_solidarite_insertion",
                                "Lieu résidence": "lieu_residence"}, inplace=True)
```

```python
#renomme les colonnes de la base de données des villes pour les rendre compatibles avec la base de données de la CAF
dt_ville.rename(columns={"department_name": "nom_dep",
                         "department_number": "num_dep",
                         "region_geojson_name" : "nom_region"}, inplace=True)
```

```python
#filtrer les données de la base de données de la CAF pour ne garder que les données de la région Bourgogne-Franche-Comté
bfc_data = df_caf[df_caf["num_region"] == "27" ]
```
